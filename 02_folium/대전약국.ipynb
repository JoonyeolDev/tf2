```python
import pandas as pd
```

```python
data = pd.read_csv('fulldata_01_01_06_P_약국.csv', encoding='cp949')
```

```python
data.head()
```

```python
# 대전 지역 약국만 추출
data.소재지전체주소
```

```python
# 지역명만 잘라서 넣기
data['지역']=data['소재지전체주소'].str.split().str.get(0)
data.지역
```

```python
# 대전광역시 찾기
dj = data[data['지역']=='대전광역시']
dj
```

```python
dj.reset_index(drop=True,inplace=True) # 인덱스 리셋
dj.head(2)
```

```python
# '구','동'으로 분리
dj['구']=dj['소재지전체주소'].str.split().str.get(1).str.strip()
dj['동']=dj['소재지전체주소'].str.split().str.get(2).str.strip()
if len(dj['구']) > 3:
    dj['구'] = dj['구'].str.slice(0,2)

```

```python
dj.구
```

```python
dj.동
```

```python
dj_gu = dj.value_counts('구')
df_gu = dj_gu.to_frame()
df_gu = df_gu.reset_index()
df_gu.columns = ['구','약국수']
df_gu = sorted(df_gu['구'])
df_gu
```

```python
jsonfile = open('02_folium/sigu.json','r',encoding='utf-8').read()
```

```python
import json
jsondata = json.loads(jsonfile)
```

```python
import json
jsondata_dj = {"type": "FeatureCollection"}
jsondata_pick = []
jsondata_dong = []

jsondata_dj['features'] = jsondata_pick
jsondata_dj # 아이디 있어야함 매칭이 되어야함

for item in jsondata['features']:# 들어오는 단위 딕셔너리 하나
    # print(item["properties"]["sidonm"])
    if item["properties"]["sidonm"] == '대전광역시':
        #print(item["properties"]["sggnm"])
        dong = item["properties"]["sggnm"].split()[-1].strip()
        item['id'] = dong
        jsondata_dong.append(dong)
        jsondata_pick.append(item)
```

```python
jsonfile = open('HangJeongDong_ver20220401.geojson.txt','r',encoding='utf-8').read()
```

```python
import json
jsondata = json.loads(jsonfile)
```

```python
import json
jsondata_dj = {"type": "FeatureCollection"}
jsondata_pick = []
jsondata_dong = []

jsondata_dj['features'] = jsondata_pick
jsondata_dj # 아이디 있어야함 매칭이 되어야함

for item in jsondata['features']:# 들어오는 단위 딕셔너리 하나
    # print(item["properties"]["sidonm"])
    if item["properties"]["sidonm"] == '대전광역시':
        #print(item["properties"]["adm_nm"])
        dong = item["properties"]["adm_nm"].split()[-1].strip()
        item['id'] = dong
        jsondata_dong.append(dong)
        jsondata_pick.append(item)
```

```python

import folium
busan_loc = [35.2144654,128.9805666]
m = folium.Map(busan_loc,zoom_start=11)
folium.Choropleth(geo_data = jsondata_dong,
                  data = dj_gu,
                  columns=['동','약국수'],
                  fill_color='YlGn',
                  key_on = 'feature.id').add_to(m)
m
```
