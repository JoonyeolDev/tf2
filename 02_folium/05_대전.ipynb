```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
# 데이터 준비하기
data = pd.read_csv('전국초중등학교위치표준데이터.csv',encoding='cp949')
```

```python
# 데이터 전처리 대전만 추려내기

# 1) 주소의 젤 첫번째를 나눠서 '지역'에 담기
data['지역'] = data['소재지지번주소'].str.split(' ').str.get(0)

# 2) '지역'이 '세종특별자치시'인 놈을 찾아서 dj변수에 넣기

sj = data[data['지역']=='제주특']

# 3) 새로운인덱스 부여
sj.reset_index(drop=True,inplace=True) #새로운 인덱스 부여
sj.head(2)

# 4) 소재지지번주소 > 구동
sj['구'] = sj['소재지지번주소'].str.split().str.get(1).str.strip()
sj['동'] = sj['소재지지번주소'].str.split().str.get(2).str.strip()

sj
```

```python
# 5) 동단위로 학교 갯수 구하기
dj_gu = dj.value_counts('동') #df.rename(columns={'A': 'a'}, index={'ONE': 'one'})
df_gu = dj_gu.to_frame() #시리즈를 df로 만들어줌
df_gu = df_gu.reset_index()

df_gu.columns = ['동', '학교명']

df_gu = sorted(df_gu['동'])
df_gu

```

```python
import json
jsonfile = open('HangJeongDong_ver20220401.geojson.txt','r',encoding='utf-8').read()
jsondata = json.loads(jsonfile)

jsondata_df_gu = {"type": "FeatureCollection"}
jsondata_pick = []
jsondata_dong = []

for item in jsondata['features']:
    # print(item['properties']['sidonm'])
    if item['properties']['sidonm'] == '대전광역시':
        # print(item['properties']['adm_nm'])
        dong = item['properties']['adm_nm'].split()[-1].strip() 
        # if dong == '가야제1동':
        #     dong = '가야1동'
        item['id'] = dong
        jsondata_dong.append(dong)
        jsondata_pick.append(item)
        
jsondata_dj['features'] = jsondata_pick
jsondata_dj

import folium
dj_loc = [36.3411098,127.3894096]
m = folium.Map(dj_loc,zoom_start=11)
folium.Choropleth(geo_data = jsondata_dj,
                  data = df_gu,
                  columns=['동','학교수'],
                  fill_color='YlGn',
                  key_on = 'feature.id').add_to(m)
m
```
