```python
# 컬럼 삭제
# 주소 전처리

data.drop(columns=['Unnamed: 8'],inplace=True) #inplace 를 사용해야 덮어쓰기 가능
```

```python
data['주소']=data['주소'].str.split('(').str.get(0)
```

```python
data.head(1)
```

```python
# 주소를 이용해서 카카오 로컬 키워드 검색을 이용해서 '위치'컬럼에 [y,x] 저장
# 함수로 작성, aaply를 이용해서 적용
# 결과값이 없을경우 '결과없음'으로 처리

def get_yx(query):
    import requests
    REST_API_KEY = 'fece9b208964c06cd0adb8ff1e570ba5'
    headers = {'Authorization': f'KakaoAK {REST_API_KEY}'}
    category_group_code='PO3'

    url = f'https://dapi.kakao.com/v2/local/search/keyword.json?query={query}&category_group_code={category_group_code}'

    r = requests.get(url, headers=headers)
    addr = r.json()
    try:
        return [addr['documents'][0]['y'],addr['documents'][0]['x']]
    except:
        return '결과없음'
```

```python
data['위치'] = data['주소'].apply(get_yx)
```

```python
data.sample(2)
```

```python
# '위치'컬럼의 값을 이용해서 마커를 생성하세요
# tooltip으로 '읍면동' 컬럼의 값을 사용해서 지도에 표시합니다.
import folium
from folium.plugins import MarkerCluster
m = folium.Map(location=[36.3279173,127.4444278],zoom_start=7)
marker_cluster = MarkerCluster().add_to(m)

for index,row in data.iterrows():
    if row['위치'] !='결과없음':
        folium.Marker(row['위치'],tooltip=row['읍면동']).add_to(marker_cluster)
m
```

```python
def jeju_map(data):

    import folium
    from folium.plugins import MarkerCluster
    m = folium.Map(location=[36.3279173,127.4444278],zoom_start=7)
    marker_cluster = MarkerCluster().add_to(m)

    for index,row in data.iterrows():
        if row['위치'] !='결과없음':
            folium.Marker(row['위치'],tooltip=row['읍면동']).add_to(marker_cluster)

```
