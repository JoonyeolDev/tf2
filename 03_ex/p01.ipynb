```python
df_last['분양가격(㎡)'] # NaN값이 있으나 NaN은 실수취급이므로 추가적인 문제가 있을것이라 예상
```

```python
pd.to_numeric(df_last['분양가격(㎡)'],errors='coerce').isna().sum()
# to_numeric은 float64 or int64 타입으로 변환시켜줌 근데 무지성 변환아니고
# to_numeric의 errors는 'raise', 'ignore', 'coerce' 3가지 옵션이 있는데
# 'raise'는 알람을 띄우며 작동중지
# 'ignore'는 알람을 무시하며 진행
# 'coerce'는 error 발생 시 NaN값으로 바꾸고 진행한다
# 근데 뒤에 isna()를 사용해 바꾼 NaN값을 삭제
```

```python
df_last['분양가격'] = pd.to_numeric(df_last['분양가격(㎡)'],errors='coerce') #뒤에 제곱미터 거슬리니깐 삭제
```

```python
df_last.info() # 분양가격(㎡) NaN값이 있기때문에 int가 아닌 float 바뀐것 확인
```

```python
df_last['평당분양가격'] = df_last['분양가격']*3.3 # 평당분양가격으로 변경
```
